```python
import requests
from bs4 import BeautifulSoup as soup
import pandas as pd
from pandas import DataFrame
```

```python
url = "https://webscraper.io/test-sites/e-commerce/allinone/computers/laptops"
result = requests.get(url)
doc = soup(result.text, "lxml")
items = doc.find_all('div', {'class': 'col-sm-4 col-lg-4 col-md-4'})
data1 = []
```

```python
for item in items:
    data1.append({
        'Product' : item.a['title'],
        'price' : item.find('h4', {'class': 'pull-right price'}).string
    })
```

```python
df = pd.DataFrame(data=data1)
```

```python
df
```

```python
price_value = df['price'].str.extract('(?P<price_value>\d+)',expand=False)
df['price_value'] = price_value.astype('int')
```

```python
price_value
```

```python
df
```

```python
df_new = df.drop(['price'],axis=1)
df_new
```

```python
df_final = df_new.rename(columns={'price_value':'Price ($)'})
df_final
```
